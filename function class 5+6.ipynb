```python
def add (a,b):
    return a+b

def subtract (a,b):
    return a-b

def multiply (a,b):
    return a*b

def divide (a,b):
    return a/b 

def cal():
    
    a=int(input("Enter a="))
    b=int(input("Enter b="))
    o= input ("Enter + - * /")
    
    if o == "+":
        print(add (a,b))
    elif o == "-":
        print(subtract(a,b))
    elif o == "*":
        print(multiply (a,b))
    elif o == "/":
        print(divide (a,b))
    else:
        print("invalid operator")
        
cal()
```

```python
def add (a,b):
    return a+b

def subtract (a,b):
    return a-b

def multiply (a,b):
    return a*b

def divide (a,b):
    return a/b 

def cal(a,b,o):
        
    if o == "+":
        print(add (a,b))
    elif o == "-":
        print(subtract(a,b))
    elif o == "*":
        print(multiply (a,b))
    elif o == "/":
        print(divide (a,b))
    else:
        print("invalid operator")

a=int(input("Enter a="))
b=int(input("Enter b="))
o= input ("Enter + - * /")

        
cal(a,b,o)
```

```python
def add (a,b):
    return a+b

def subtract (a,b):
    return a-b

def multiply (a,b):
    return a*b

def divide (a,b):
    return a/b 

def cal():
    
    a=int(input("Enter a="))
    b=int(input("Enter b="))
    o= input ("Enter + - * /")
    
    if o == "+":
        print(add (a,b))
    elif o == "-":
        print(subtract(a,b))
    elif o == "*":
        print(multiply (a,b))
    elif o == "/":
        print(divide (a,b))
    else:
        print("invalid operator")
        
    x= input("Enter y for more calculation")
    if x == 'y':
        cal()
        
cal()
```

```python
s = 0
n = int(input("Enter n="))
def summ():
    global s
    for i in range(n):
        x = int(input("Enter x="))
        s = s+x
        
summ()
print(s)
```
